**Write a NumPy program to implement following operation**

**1) To convert a list of numeric values into a one-dimensional NumPy
array**

```python
import numpy as np
```

```python
l = [23,8,18,29]
print("original list :",l)

a = np.array(l)
print("one-dimensional NumPy array :",a)
```

**2) To create a 3x3 matrix with values ranging from 2 to 10**

```python
a2 = np.arange(2,11).reshape(3,3)
print("A2 :",a2)
```

**3) To append values at the end of an array**


```python
a3 = np.append(a, [12,11])
print('Array after appending : ', a3)
```

**4) To create another shape from an array without changing its data(3x2
to 2x3)**

```python
a3 = np.array([[1,2],
               [3,4],
               [5,6]])
print("A3 :",a4)
print("\n Shape of A3 :",a4.shape)

a4 = a3.reshape(2,3)
print("\n A4 :",a4)
print("\n Shape of A4 :",a4.shape)
```
