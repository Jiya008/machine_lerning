* **Split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order**

```python

a = np.array([4,6,8,10,11,12,18,20,23,24,25,29])

# Displaying the original array 'x'
print("\nOriginal array : \n", a)

# Splitting the array 'x' into sub-arrays
a1, a2, a3 = np.split(a, [2,6])
print("\nAfter splitting : \n", a1, a2, a3)
```
