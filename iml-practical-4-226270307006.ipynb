**3. To calculate mean across dimension, in a 2D numpy array**

```python
a = np.array([[23,6], [29,18]])

print("A  :", a)

print("mean of a :",np.mean(a))
print("mean of a using axis=0 : ",np.mean(a, axis=0))
print("mean of a using axis=1 : ",np.mean(a, axis=1))
```
